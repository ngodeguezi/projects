```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://lista.mercadolivre.com.br/'
```

```python
produto = input('Qual produto você deseja? ')
```

```python
response = requests.get(url + produto)
```

```python
site = BeautifulSoup(response.text, 'html.parser')
```

```python
print(site.prettify())
```

```python
produto_site = site.find('div', attrs={'class': 'andes-card andes-card--flat andes-card--default ui-search-result ui-search-result--core andes-card--padding-default'})
```

```python
print(produto_site.prettify())
```

```python
descricao = produto_site.find('h2', attrs={'class': 'ui-search-item__title'})
```

```python
descricao.text
```

```python
link = produto_site.find('a', attrs={'class': 'ui-search-link'})
```

```python
link['href']
```

```python
preço_simbolo = produto_site.find('span', attrs={'price-tag-symbol'})
preço_real = produto_site.find('span', attrs={'price-tag-fraction'})
preço_centavos = produto_site.find('span', attrs={'price-tag-cents'})
```

```python
print(preço_simbolo.text + preço_real.text + ',' + preço_centavos.text)
```

```python
produtos_site = site.findAll('div', attrs={'class': 'andes-card andes-card--flat andes-card--default ui-search-result ui-search-result--core andes-card--padding-default'})
```

```python
# Armazenando todos os produtos e seus links em um arquivo xlsx.

lista = []

for produto_site in produtos_site:
    
    descricao = produto_site.find('h2', attrs={'class': 'ui-search-item__title'})
    
    link = produto_site.find('a', attrs={'class': 'ui-search-link'})
    
    preço_simbolo = produto_site.find('span', attrs={'price-tag-symbol'})
    preço_real = produto_site.find('span', attrs={'price-tag-fraction'})
    preço_centavos = produto_site.find('span', attrs={'price-tag-cents'})
    
    if (preço_centavos):
        lista.append([descricao.text, preço_simbolo.text + preço_real.text + ',' + preço_centavos.text, link['href']])
    else:
        lista.append([descricao.text, preço_simbolo.text + preço_real.text, link['href']])
        
lista_produtos = pd.DataFrame(lista, columns=['Descrição', 'Preço', 'Link'])
        
lista_produtos.to_excel('produtos.xlsx', index=False)
```
